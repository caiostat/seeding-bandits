## Measure: Reaction Probability

<font size = 4>The reaction probability is the probability of receiving a reaction in up to 7 days. It is calculated as follows:<br>
<br>1. For each action from a creator towards a fan, we look for a reaction from the same fan to that creator, the reaction must always happen after the action took place.<br>
<br>2. If the reaction happened less than 7 days after the action, the action is flagged as successful (by means of an indicator variable).
<br>3. We calculate the mean reaction probability across dyads and weeks by averaging the `successful` flag (on user_id, fan_id and week).
<br>4. Finally we average the measure obtained in step 3 across users and weeks.<br> 
<br>   Notes: 
<br>1. one reaction cannot make 2 actions successful. This is prevented by indexing of reactions and keeping only one occurence of each index (we keep the one that is closer to the action).<br>
    2. If two actions were done before on reaction is obtained, even if the reaction happened within the 7 days window from the first action, only the second one will be flagged as successful.
</font>

```python
def process_date(date):
    year = date[0:4]
    week = date[6:]
    day = "1"
    date = "{}-{}-1".format(year, week)
    dt = datetime.strptime(date, "%Y-%W-%w")
    return dt
```

<font size = 4>
Goal: to diagnose how active a creator’s dyads are, is in terms of the dyad valence. The valence is defined as the difference between actions started – actions received (reactions) per dyad.<br>
Reaction valence is the number of reactions minus the number of actions.
</font>
